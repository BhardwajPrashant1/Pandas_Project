```python
import pandas as pd
```

```python
#Read Dataframe 1
df1=pd.read_csv(r"https://raw.githubusercontent.com/AlexTheAnalyst/PandasYouTubeSeries/main/LOTR.csv")
df1.head()
```

```python
#Read Dataframe2
df2=pd.read_csv("https://raw.githubusercontent.com/AlexTheAnalyst/PandasYouTubeSeries/main/LOTR%202.csv")
df2.head()
```

```python
#Merge both dataframes on single column:Inner returns only the matching rows from both DataFrames
pd.merge(df1,df2, on = "FellowshipID",how='inner')
```

```python
#Merge both dataframes on multiple columns
pd.merge(df1,df2, on = ["FellowshipID","FirstName"],how='inner')
```

```python
#Alternate way to merge both
df1.merge(df2,on=["FellowshipID","FirstName"],how='inner')
```

```python
#try outer merging:includes all the rows from both dataframes
df1.merge(df2,on=["FellowshipID","FirstName"],how='outer')
```

```python
#left merge:returns all the rows from the left DataFrame and only the matching rows from the right DataFrame
df1.merge(df2,on=["FellowshipID","FirstName"],how='left')
```

```python
#right merge:returning all the rows from the right DataFrame and only the matching rows from the left DataFrame
df1.merge(df2,on=["FellowshipID","FirstName"],how='right')
```

```python
#Cross Join:combine every row from the first dataframe with every row from the second dataframe
df1.merge(df2,how='cross')
```

```python
#Concat along rows axis=0:by default Value
pd.concat([df1,df2], join = 'outer', axis = 0)
```

```python
#In the above line original Index is preserved suppose we ant to set new index
pd.concat([df1,df2], join = 'outer', axis =0).reset_index(drop='True')
```

```python
#Concat along Column axis=1
pd.concat([df1,df2], join = 'outer', axis = 1)
```
