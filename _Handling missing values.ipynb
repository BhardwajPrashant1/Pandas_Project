```python
#Get the missing value count for top 10 rows
df.isna().sum().head(10)
```

```python
#Get the percentage of data missing in top 10 columns Way:1
total_values=df.iloc[:,:10].shape[0]
missing_values = df.iloc[:,:10].isna().sum()
percent_missing = (missing_values / total_values) * 100
percent_missing
```

```python
#Get the percentage of data missing in top 10 columns Way:2
df.iloc[:,:10].isna().mean()*100
```

```python
#remove all the rows that contain a missing value
df.dropna()
#look's like we have removed all the data as  every row in our dataset had at least one missing value 
```

```python
#remove all columns with at least one missing value
df1=df.dropna(axis=1)
df1.head()
```

```python
#How many colums we have removed as compared to our original Datast
print(df.shape[1])
print(df1.shape[1])
```
