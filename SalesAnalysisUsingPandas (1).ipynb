# What city had highest number of Sales?

```python
df.head()
```

```python
#First Add a city column
```

```python
df["CityName"]=df["Purchase Address"].str.split(",").str.get(1)
```

```python
df.head()
```

```python
#Another way of doing this is by using .apply()
```

```python

df['City']=df["Purchase Address"].apply(lambda x:x.split(',')[1])
```

```python
#Making it a function
def get_city(address):
    return address.split(',')[1]

df['City']=df["Purchase Address"].apply(lambda x:get_city(x))
```

```python
df.drop(columns='City',inplace=True)
```

```python
df
```

```python
def get_state(address):
    return address.split(',')[2].split(' ')[1] ##get rid of Zop Code

df['State_City']=df["Purchase Address"].apply(lambda x :get_state(x))
```

```python
df
```

```python
df["Address"]=df['CityName']+df['State_City']
```

```python
df
```

```python
df["Address"]=df['CityName']+ ' (' + df['State_City'] + ')' 
```

```python
df.head()
```

#What city had highest number of Sales?

```python
CitywithMaxSales=df.groupby('Address').sum()
CitywithMaxSales
```

```python
cities= [address for address,df in df.groupby('Address')]
plt.bar(cities,CitywithMaxSales['Sales'])
plt.xticks(cities,rotation='vertical')
plt.xlabel('CityNames')
plt.ylabel('Sales in k')
plt.show()
```
